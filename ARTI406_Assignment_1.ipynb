# Assignment 1: Exploratory Data Analysis (EDA)

EDA is the first and most important step in any Machine Learning project.
Before building models, we must understand:

- What does the data represent?
- Are there missing values?
- Are there outliers?
- What patterns exist?
- Which variables influence others?

If we do not understand the data, we cannot build a good model.

### Why EDA is Important

In real-world machine learning projects:

    1- 70–80% of the time is spent on understanding and cleaning data
    2- Only 20–30% is spent building models

**EDA helps us:**

- Detect errors
- Identify trends
- Discover relationships
- Make business decisions

## Introduction:
This project performs Exploratory Data Analysis (EDA) on a Sleep Health and Lifestyle dataset...

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
# Load Dataset
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
# Display first 5 rows. If you want to display all rows, use df.head(-1)
df.head(-1)
```

## Data Cleaning
### Data Before cleaning

```python
df.info()
df.isnull().sum()
```

### Data After cleaning

```python
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None')
df.isnull().sum()
```

### Check Missing Values

```python
# Check all values
print(df.isna())
```

```python
print(df.isna().sum())
```

### Check duplicate rows

```python
# checking duplicate rows
df.duplicated()[df.duplicated()==True]
```

### No. of rows and columns

```python
# finding number of rows and columns

print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

### Data type of columns

```python
# viewing the data types of columns
df.dtypes
```

```python
# Fill missing values
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None')

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Sleep Duration'] = pd.to_numeric(df['Sleep Duration'], errors='coerce')
df['Quality of Sleep'] = pd.to_numeric(df['Quality of Sleep'], errors='coerce')
df['Stress Level'] = pd.to_numeric(df['Stress Level'], errors='coerce')
df['Daily Steps'] = pd.to_numeric(df['Daily Steps'], errors='coerce')

df.dtypes
```

```python
# Statistical summary
df.describe(include='all')
```

### Descriptive summary Statistics

Distribution of Sleep Duration

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Sleep Duration"], bins=20, kde=True)
plt.title("Distribution of Sleep Duration")
plt.xlabel("Sleep Duration (Hours)")
plt.ylabel("Frequency")
plt.show()
```

Insight:
Most individuals sleep between 6 and 8 hours, indicating that the majority maintain a relatively normal sleep duration.

- Shows how shipment sizes are distributed
- Right skew = many small shipments, few large ones

Distribution of Quality of Sleep

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Quality of Sleep'], bins=10, kde=True)
plt.title("Distribution of Quality of Sleep")
plt.xlabel("Quality of Sleep")
plt.ylabel("Frequency")
plt.show()
```

Insight:
Sleep quality is generally moderate to high, suggesting that most individuals experience acceptable sleep conditions.


### Average Sleep Duration by *Occupation*

```python
occupation_sleep = df.groupby('Occupation')['Sleep Duration'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
occupation_sleep.plot(kind='bar')
plt.title("Average Sleep Duration by Occupation")
plt.ylabel("Average Sleep Duration (Hours)")
plt.xlabel("Occupation")
plt.xticks(rotation=45)
plt.show()

occupation_sleep
```

Insight:
Sleep duration varies across different occupations, indicating that job type may influence sleeping habits and schedules.

### Quality of Sleep by BMI Category

```python
bmi_sleep = df.groupby('BMI Category')['Quality of Sleep'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
bmi_sleep.plot(kind='bar')
plt.title("Average Quality of Sleep by BMI Category")
plt.ylabel("Average Quality of Sleep")
plt.xlabel("BMI Category")
plt.xticks(rotation=30)
plt.show()

bmi_sleep
```

Insight:
BMI categories show differences in sleep quality, suggesting that body condition may affect how well individuals sleep.

### Number of People by Occupation

```python
occupation_count = df['Occupation'].value_counts()

plt.figure(figsize=(10,5))
occupation_count.plot(kind='bar')
plt.title("Number of People by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

occupation_count.head(10)
```

Insight:
Some occupations such as nurses and doctors appear more frequently in the dataset, while others have fewer individuals.

### Stress Level vs Sleep Duration

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Stress Level', y='Sleep Duration', data=df)
plt.title("Stress Level vs Sleep Duration")
plt.xlabel("Stress Level")
plt.ylabel("Sleep Duration (Hours)")
plt.show()
```

Insight:
Higher stress levels are generally associated with slightly lower sleep duration, indicating that stress negatively impacts sleep.

### Correlation Matrix

```python
plt.figure(figsize=(8,6))
numeric_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='Blues')
plt.title("Correlation Matrix")
plt.show()
```

Insight:
The correlation matrix shows that stress level has a negative relationship with sleep quality, while physical activity and sleep duration have a mild positive relationship.

- Correlation close to 1 = strong positive relationship
- Close to 0 = weak relationship

### Average Sleep Duration by Age

```python
sleep_by_age = df.groupby('Age')['Sleep Duration'].mean()

plt.figure(figsize=(10,5))
sleep_by_age.plot()
plt.title("Average Sleep Duration by Age")
plt.xlabel("Age")
plt.ylabel("Average Sleep Duration (Hours)")
plt.show()
```

Insight:
Sleep duration shows noticeable variation across age groups, with older individuals generally experiencing slightly longer sleep duration compared to younger groups.
